```python
import json
with open('sectors3.json') as f:
    sectors3 = json.load(f)
print(f"Loaded {len(sectors3)} sectors")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract ward counts
ward_counts = [count for _, count in sectors3]

# Create histogram
plt.figure(figsize=(12, 6))
plt.hist(ward_counts, bins=range(1, max(ward_counts) + 2), align='left', rwidth=0.8)
plt.title('Distribution of Ward Counts per Sector')
plt.xlabel('Number of Wards')
plt.ylabel('Number of Sectors')
plt.grid(True, alpha=0.3)

# Add some statistics
mean_wards = np.mean(ward_counts)
median_wards = np.median(ward_counts)
plt.axvline(mean_wards, color='red', linestyle='dashed', alpha=0.5, label=f'Mean: {mean_wards:.1f}')
plt.axvline(median_wards, color='green', linestyle='dashed', alpha=0.5, label=f'Median: {median_wards:.1f}')
plt.legend()

plt.show()

# Print summary statistics
print(f"Summary Statistics:")
print(f"Mean wards per sector: {mean_wards:.2f}")
print(f"Median wards per sector: {median_wards:.2f}")
print(f"Min wards: {min(ward_counts)}")
print(f"Max wards: {max(ward_counts)}")
print(f"Standard deviation: {np.std(ward_counts):.2f}")
```
